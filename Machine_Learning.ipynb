```python
import pandas as pd
data1=pd.read_csv('/content/googl_data_2020_2025.csv')
```

```python
data1.isnull().sum()
```

```python
data1['High']=data1['High'].fillna(data1['High'].mode())
```

```python
data1.isnull().sum()
```

```python
data1['Close']=data1['Close'].fillna(data1['Close'].mode())
data1['Low']=data1['Low'].fillna(data1['Low'].mode())
data1['Open']=data1['Open'].fillna(data1['Open'].mode())
data1['Volume']=data1['Volume'].fillna(data1['Volume'].mode())
```

```python
data1.isnull().sum()
```

```python
data1['Adj Close']=data1['Adj Close'].fillna(data1['Adj Close'].mode())
```

```python
data1.isnull().sum()
```

```python
data1=data1.dropna()
```

```python
data1.isnull().sum()
```
